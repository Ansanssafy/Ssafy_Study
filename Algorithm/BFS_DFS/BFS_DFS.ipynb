## **2.  **

##### **1. DFS **
```python
# DFS  
def dfs(graph, v, visited):
    #    
    visited[v] = True
    print(v, end=' ')
    #       
    for i in graph[v]:
        if not visited[i]:
            dfs(graph, i, visited)

#      (2 )
graph = [
    [],
    [2, 3, 8],
    [1, 7],
    [1, 4, 5],
    [3, 5],
    [3, 4],
    [7],
    [2, 6, 8],
    [1, 7]
]

#      (1 )
visited = [False] * 9

#  DFS  
dfs(graph, 1, visited)

```


##### **2. BFS **
```python
from collections import deque

# BFS  
def bfs(graph, start, visited):
    #    deque  
    queue = deque([start])
    #    
    visited[start] = True
    #    
    while queue:
        #     
        v = queue.popleft()
        print(v, end=' ')
        #       
        for i in graph[v]:
            if not visited[i]:
                queue.append(i)
                visited[i] = True

#      (2 )
graph = [
    [],
    [2, 3, 8],
    [1, 7],
    [1, 4, 5],
    [3, 5],
    [3, 4],
    [7],
    [2, 6, 8],
    [1, 7]
]

#      (1 )
visited = [False] * 9

#  BFS  
bfs(graph, 1, visited)
```

### * python  - DFS
```python
import sys

# DFS       
def dfs(x, y):
    #      
    if x <= -1 or x >= n or y <= -1 or y >= m:
        return False
    #     
    if graph[x][y] == 0:
        #    
        graph[x][y] = 1
        # , , ,     
        dfs(x - 1, y)
        dfs(x, y - 1)
        dfs(x + 1, y)
        dfs(x, y + 1)
        return True
    return False

# N, M     
n, m = map(int, sys.stdin.readline().rstrip().split())

# 2     
graph = []
for i in range(n):
    graph.append(list(map(int, input())))
    
#  ()   
result = 0
for i in range(n):
    for j in range(m):
        #   DFS 
        if dfs(i, j) == True:
             result += 1

print(result) #  
```

### * python  - BFS
```python
import sys
N_computer = int(input())
visited = [False]*(N_computer+1)
N_E = int(input())
E=[[] for i in range(N_computer+1)]

for i in range(N_E):
    S,D = map(int, input().split())
    E[S].append(D)
    E[D].append(S)
count = 0

from collections import deque
def bfs(E,v,visited):
    queue = deque([v])  #  .
    visited[v]=True   # 
    global count
    while queue: 
        x = queue.popleft()      #       
        for i in E[x]:
            if not visited[i]:
                queue.append(i)
                visited[i]=True 
                count+=1
bfs(E,1,visited)
print(count)
```
