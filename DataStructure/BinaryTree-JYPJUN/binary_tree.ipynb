```python
class Node:                             #    
    def __init__(self, value):
        self.middle = value
        self.left = None                #     None
        self.right = None               #     None

class BinarySearchTree:
    def __init__(self):                 #  root None 
        self.root = None
    
    def bst_insert(self, number):       # insert
        new_node = Node(number)         #     
        
        if self.root is None:           #   None
            self.root = new_node        #      
            return                      
        
        current = self.root             #  None     root 
        
        while True:                     # insert   None     
            parent = current
            if number < current.middle:
                current = current.left  # current Node     
                if current is None:
                    parent.left = new_node
                    break
            else:
                current = current.right
                if current is None:
                    parent.right = new_node # current Node     
                    break
    
    def bst_search(self, search):        # search
        current = self.root              # root  
        height = 0
        while True:             
            if current is None:          # Tree   None
                return None
            
            if current.middle > search:  #  
                current = current.left   #     
            elif current.middle < search:#  
                current = current.right  #     
            else: #   return
                return f'  Tree height : {height}  . '
            height += 1
    
    #  
    def preorder(self, node):
        if node is not None:
            print(node.middle, end=' ')
            self.preorder(node.left)
            self.preorder(node.right)
            
    #  
    def inorder(self, node):
        if node is not None:
            self.inorder(node.left)
            print(node.middle, end=' ')
            self.inorder(node.right)
            
    #  
    def postorder(self, node):
        if node is not None:
            self.postorder(node.left)
            self.postorder(node.right)
            print(node.middle, end=' ')
            
# ------------------------------------------------------
lst = BinarySearchTree()

lst.bst_insert(5)
lst.bst_insert(10)
lst.bst_insert(2)
lst.bst_insert(4)
lst.bst_insert(1)
lst.bst_insert(13)
lst.bst_insert(9)

print(lst.bst_search(2))    #   Tree height : 1  .
print(lst.bst_search(5))    #   Tree height : 0  .
print(lst.bst_search(13))   #   Tree height : 2  .

lst.preorder(lst.root) # 5 2 1 4 10 9 13
print()
lst.inorder(lst.root)  # 1 2 4 5 9 10 13
print()
lst.postorder(lst.root) # 1 4 2 9 13 10 5
```
