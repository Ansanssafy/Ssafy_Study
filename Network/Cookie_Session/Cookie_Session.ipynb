## **1. HTTP**
   
- 쿠키와 세션의 개념이 생겨난 이유는 주로 HTTP 프로토콜의 본질적인 한계로 비롯된다.

- HTTP는 기본적으로 상태가 없는(Stateless) 프로토콜이기 때문에, 클라이언트와 서버 간의 각 요청은   
독립적으로 처리되며, 이전의 요청이나 이후의 요청과 연관되지 않는다. 따라서 아래와 같은 문제가 발생한다.

    

#### 1. 상태 유지 불가(Connectless, Stateless)

- HTTP는 Connectless 방식으로 작동한다. 서버에 연결, 요청, 응답을 하면 연결을 끊어버린다.
    - 기본적으로 자원 하나에 대해서 하나의 연결을 만듦

- 장점 : 불특정 다수를 대상으로 하는 서비스에 적합
- 단점 : 연결을 끊어버리기 때문에, 클라이언트의 이전 상태를 알 수 없다. 이러한 특징을 stateless라고 부르며, Connectless에서 파생되는 특징이다.

#### 2. 사용자 식별의 어려움

- HTTP 요청 간에 클라이언트를 식별할 방법이 없기 때문에, 서버는 각각의 요청이 동일한 사용자인지 구분할 수 없다.

    

- 일반적인 요청 헤더
    - Host : 요청을 보내는 대상 서버의 호스트명
    - User-Agent : 클라이언트 소프트웨어 정보
    - Accept : 클라이언트가 받아들일 수 있는 미디어 타입
    - Accept-Language : 클라이언트가 선호하는 언어
    - Referer : 현재 요청의 출처가 되는 URL
    - Cookie : 클라이언트가 이전에 받은 쿠키 데이터 (일단 제외,,)

1. 일반적으로 각 요청이 독립적으로 처리되므로 서버는 클라이언트의 이전 상태나 요청을 기억하지 않는다.

2. 헤더의 비유일성 : HTTP 헤더는 클라이언트의 특성을 설명하지만, 고유하게 식별할 수 있는 정보는 포함되지 않는다.

## **2. Cookie**


- 쿠키는 클라이언트(브라우저) 로컬에 저장되는 Key-Value 쌍의 작은 데이터 파일로, 웹 사이트를 접속할 때 생성되는 정보를 담은 임시파일이다.

    

- 웹 서버는 HTTP 응답에 쿠키를 포함하여 클라이언트에게 보낼 수 있으며, 이후 클라이언트는 해당 쿠키를 HTTP 요청에 포함시켜 서버에 다시 보낸다.


#### 1. 쿠키의 특징

- Cookie   


- 쿠키 이름, 쿠키값, 만료 시간, 전송할 도메인명, 전송할 경로, 보안연결 여부, HttpOnly여부 등으로 구성
- 최대 4KB까지 저장 가능
- 서버 데이터를 공유하는 용도로 서버 요청 시 HTTP 요청 Headers에 전달됨
- 데이터 유효기간 지정 가능
- 파일에서 읽기 때문에 세션보다 상대적으로 빠름

#### 2. 쿠키의 종류

1. 만료기간 별 종류

    

2. 도메인 별 종류

    


#### 3. 쿠키의 단점

- HTTP 요청 시 자동으로 모든 쿠키가 전송되어 불필요한 트래픽 발생

- 방문했던 웹 사이트에 대한 정보 및 개인정보가 기록되어 남기 때문에 개인정보 유출의 위험 발생

- 서버가 가지고 있지 않고 사용자에게 저장되기 떄문에 임의로 수정이 가능하고 데이터 유출에 취약

- CSRF, XSS 등의 보안적 문제가 있음

    - CSRF : 사용자 권한을 이용한 공격으로 사이트에 로그인 되어있는 유저에게 악성 스크립트 실행

        

    - XSS : token을 통해 사용자의 민감한 정보를 탈취

## **3. Session**

- Session   


- 세션은 서버 측에서 사용자 상태를 유지하기 위해 도입되었다.

    

- 세션의 도입은 주로 서버 측에서 사용자 상태를 관리하고, 더 많은 데이터와 복잡한 상태 정보를 안전하기 유지하기 위함

- 로그인 상태 유지, 사용자별 데이터 관리 등의 기능을 구현하기 위해 세션 사용


#### 1. Session의 특징

- 세션 ID는 웹 브라우저 당 1개씩 생성되어 웹 컨테이너에 저장되는 방식이기 떄문에 브라우저 종료 시 소멸

- 같은 브라우저 탭 안에서는 새로고침을 해도 유지

- 단발성 정보를 저장하기에 적합(상품 조회 조건 등)

- 서버에 저장되므로 쿠키보다 보안 측면에서 뛰어남

- 제한 시간을 적용해 일정 시간 응답이 없으면 세션을 제거하도록 설정 가능

#### 2. Session의 단점

- 사용자가 계속해서 늘어날수록 서버 메모리를 많이 필요로 하기 때문에 서버 과부하로 인한 성능 저하 문제점 발생

- 서버에서 처리하기 때문에 쿠키에 비해 비교적 느림

- 보안성 측면에서 Cookie보다 뛰어나지만 Session ID를 탈취하여 클라이언트인 척 위장 후 로그인하여 사용할 수 있다는 한계가 존재함

    - Session Hijacking
