## **1. 데이터베이스에 대한 기초 지식**

#### 1. 데이터베이스(Database, DB)란?   
   
- 데이터베이스는 '데이터의 집합'이라고 정의할 수 있음
- 기업에서 사용되는 모든 정보들이 모두 데이터베이스에 기록된다.
- 이 데이터베이스를 관리하고 운영하는 소프트웨어를 DBMS(Database Management System)이라고 한다.

#### 2. DBMS의 분류
- DBMS는 계층형(Hierarchical), 망형(Network), 관계형(Relational), 객체지향형(Object-Oriented), 객체관계형(Object-Relational) 등으로 분류된다.
- 현재 사용되는 DBMS는 관계형 DBMS가 가장 많은 부분을 차지하며 MYSQL도 관계형 DBMS에 포함된다.

#### 3. 관계형 DBMS   
   
- 관계형 DBMS의 데이터베이스는 테이블(table)이라는 최소 단위로 구성되며, 이 테이블은 하나 이상의 열(column)과 행(row)으로 이루어져 있다.

#### 4. SQL-DBMS에서 사용하는 언어   
   
- DBMS는 다양한 프로그램이 존재하나 표준 SQL의 기준을 따르며, 프로그램마다 각각의 기능이 있다.

---

## **2. 정규화(Normalization) 란?**
- 관계형 데이터베이스의 설계에서 중복을 최소화하게 데이터를 구조화하는 프로세스를 정규화(Normalization)라고 한다.
- 데이터베이스 정규화의 목표는 이상이 있는 **관계를 재구성하고 작고 잘 조직된 관계를 생성**하는 것에 있다.


#### 1. 제1정규형
1. 각 컬럼이 하나의 속성만을 가져야한다.
2. 하나의 컬럼은 같은 종류나 타입(type)의 값을 가져한다.
3. 각 컬럼이 유일한 이름을 가져야한다.
4. 칼럼의 순서가 상관없어야 한다.   

   
- Q. 103번 김민수가 골프초급을 등록하면?

   
- 단점1. 골프초급 수강신청한 사람 찾기가 귀찮아짐 (WHERE 프로그램 LIKE '%골프초급%')
- 단점2. 프로그램명 수정도 어려워짐


- 위와 같이 한 칸에 하나의 데이터만 넣는 방식을 제1정규화 라고 한다.
- 제1정규화가 완료된 테이블을 제1정규형 테이블 이라고 한다.


#### 2. 제2정규형
- 테이블에 존재하는 'composite primary key'에 종속된 'partial dependency'를 제거한 테이블
- 현재 테이블의 핵심주제와 상관없는 컬럼을 다 제거
1. 제 1정규화를 만족해야한다.
2. 모든 컬럼이 부분적 종속(Partial Dependency)가 없어야한다. 모든 칼럼이 완전 함수 종속을 만족해야 한다.


- Q. 헬스 가격을 7000으로 수정하려면?
    - 해당 데이터의 값들을 일일이 찾아서 수정해야함
    - 그러나 데이터의 수가 무수히 많아지면 데이터를 전부 수정하기 어려워짐
    - 따라서 제2정규화를 해야한다.
- 제2정규화 == 현재 테이블의 주제와 관련없는 컬럼을 다른 테이블로 빼는 작업


- 장점 : 가격을 수정할 때 참조하는 테이블의 값만 바꿔주면 됨
- 단점 : 손흥민이라는 사람이 내야할 금액을 조회할 때 테이블을 여러 개 조회해야 할 수도 있다.

#### 3. 제3정규형
- 제2정규화를 진행한 테이블에 대해 이행적 종속을 없애도록 테이블을 분해한 테이블
1. 제 2정규화를 만족해야한다.
2. 기본키를 제외한 속성들 간의 이행 종속성(Transitive Dependency)이 없어야한다.

   
- composite primary key는 없고 Primary key 하나만 있음
- '출신대학' 이라는 컴럼은 Primary key와는 전혀 상관없고, '강사'라는 컬럼에 종속되어 있음
- 일반 컬럼에만 종속된 컬럼을 다른 테이블로 빼는 것이 제3정규화이다.

   
- 장점 : 데이터의 수정이 편해진다.
- 단점 : 특정 데이터를 조회할 때 테이블을 타고 조회해야하기 때문에 복잡해진다.
