# *Static & Dynamic Routing*

---

## **1. Routing**

  - Routing 기술을 초기 인터넷의 발전과 함께 네트워크에서 데이터 패킷을 효율적으로 전송하기 위해 개발됨

    

    - Routing 기술 개발의 주요 이유

      1. 네트워크 확장과 효율성
      2. 데이터 전송 비용 절감
      3. 신뢰성 및 회복성 확보
      4. 네트워크 관리의 용이성
  
   
  
  - Routing은 네트워크에서 데이터 패킷(Packet)이 출발지에서 목적지로 전달되는 경로를 결정하는 과정을 말함

  - Router는 네트워크에서 Routing을 수행하는 네트워크 장치
    - 네트워크의 경계에서 동작해 다른 네트워크와의 연결점 역할을 한다.
     


#### 1. Routing Table

- Router에 내장된 데이터베이스로, 네트워크 내의 경로 정보를 저장한다.
  
  - 목적지 네트워크에 대한 최적 경로

  - 해당 경로로 패킷을 전달하기 위해 사용해야 하는 인터페이스

  - 해당 경로의 Metric


#### 2. Metric

- Metric은 Routing 결정을 할 때 사용되는 값으로, 네트워크 경로의 효율성과 성능을 평가하는 데 사용된다.

  - 평가 요소

    | 평가요소 | 내용 |
    | --- | --- |
    | 홉 카운트 (hop count) | 패킷이 목적지에 도달하기 위해 커쳐야 하는 라우터의 수 |
    | 대역폭 (bandwidth) | 일정 시간(보통 1초) 동안 전송 가능한 데이터의 최대 용량 |
    | 지연 시간 (latency) | 패킷이 출발지에서 목적지까지 도달하는 데 걸리는 시간 |
    | 신뢰성 (Reliability) | 링크의 오류율이나 다운타임과 같은 요소를 고려하여 산출된 값 |
    | 부하 (Load) | 네트워크 연결의 현재 트래픽 양 |
    | 비용 (Cost) | 네트워크 연결에 할당된 비용 값 |

## **2. Static Routing**

- 통신이 불가능한 네트워크와 통신을 가능하게 만들기 위한 절차

  

- Static Routing은 Routing Table에 경로를 수동으로 추가해야하는 프로세스
  - =(관리자에 의해 Routing Table이 유지/관리 되는 방법)

  - 관리자는 각 목적지별로 경로를 Routing Table에 입력한다.

  - Table이 생성되면 인터넷에 변화가 있더라도 테이블의 내용이 자동적으로 갱신될 수 없다.

  - 정적 경로 설정은 경로 설정 정보를 각 Router에 고정적으로 설정하는 방법이므로,   
  라우터 사이에 경로 설정 프로콜에 따른 정보를 교환하지 않는다.


##### 1. Static Routing의 장단점

1. 장점

  - 데이터가 전달되는 경로를 관리자가 직접 제어

  - 안전성이 높음
  
2. 단점

  - 라우터가 많고 목적지가 많은 경우 설정이 어려움 (ex - 설정해야할 경로가 총 10개라면??)

  - 장애가 발생했을 경우 대처가 어려움

## **3. Dynamic Routing**

- Routing Protocol에 의해 자동으로 Routing Table을 구성하는 방법

- 자동으로 경로가 결정되는 프로토콜로 Router가 판단하여 가장 효율적인 방법으로 패킷을 전송함

  

- Dynamic Routing 종류

  1. RIP (거리 벡터 라우팅 프로토콜)

  2. OSPF (링크 상태 라우팅 프로토콜)

  3. EIGRP (향상된 거리 벡터 라우팅 프로토콜)

  4. BGP (경로 벡터 라우팅 프로토콜)

##### 1. Dynamic Routing의 장단점

1. 장점

  - 데이터가 전달되는 경로를 자동으로 결정

  - 장애가 발생할 경우 대처가 쉬움
  
2. 단점

  - 데이터가 전달되는 경로를 관리자가 제어할 수 없음

  - 안정성이 낮고 장애가 발생할 가능이 있음

##### 2. Interior Gateway Protocol and Exterior Gateway Protocol

- Dynamic Routing은 내부라우팅과 외부라우팅 2가지로 나뉜다.

  

- Autonomius System (AS)
  - AS Number(망식별번호) : 각각의 자율시스템을 식별하기 위한 인터넷 상의 고유 번호
  - 하나의 그룹/기관/회사 같이 동일한 라우팅 정책으로 하나의 관리자에 의해 운영되는 네트워크

- 내부라우팅 (IGP) : 같은 AS 내부의 라우팅 정보를 교환하는 프로토콜

- 외부라우팅 (EGP) : 다른 AS 간의 라우팅 정보를 교환(다른 AS와 연결)하는 프로토콜


##### 3. IGP

- IGP는 단일 자치 영역(AS, Autonomous System) 내부에서 라우터들 간에 라우팅 정보를 교환

- 하나의 자치 영역은 일반적으로 하나의 관리적 도메인을 나타낸다. (회사 내부 네트워크, 대학교 캠퍼스 네트워크)

- IGP는 AS 내부에서 동작하며, 라우팅 테이블을 구성하고 관리하여 네트워크 내의 최적의 경로를 결정

- 주요 IGP 프로토콜로 RIP, OSPF, IS-IS 등이 있음

##### 4. EGP

- EGP는 서로 다른 자치 영영 간에 라우팅 정보를 교환하는 데 사용되는 프로토콜

- 다른 네트워크 도메인들 사이에서 경로를 교환하여 인터넷과 같은 대규모 네트워크에서 라우팅을 지원

- EGP는 AS 간 경계 라우터들 간에 사요되며, 주유 프로토콜로 BGP가 있음
