## **1. 암호화란?**
   
   
   
1. 암호(Cryptography)란 **0과 1로 이루어진 이진수 데이터를 수학적 계산을 통한 비트 변경을 수행**하는 것을 말한다.

2. 정상적인 데이터의 2진수 데이터를 암호화라는 과정으로 데이터를 변경하고, 복호화라는 과정을 통해 데이터를 원래대로 돌리는 기술을 의미한다.

    - 암호화(Encrypt)는 사용자가 입력한 데이터를 알아볼 수 없는 데이터로 변경하는 과정
    
    - 복호화(Decrypt)는 암호화된 데이터를 정상적인 데이터로 변경하는 과정

3. 암호화에는 단방향 암호와 양방향 암호 두 가지가 존재한다.

    - 단방향 암호(해시)는 암호화를 수행하고 나온 데이터를 다시 원래의 데이터로 돌릴 수 없는 암호화 방식을 의미
        - 단방향 암호화에서 나오는 데이터는 원래 데이터의 1bit만 바뀌어도 전혀 다른 암호문으로 변하기 때문에, 보통 데이터가 변조되지 않았음을 나타낼 때 사용

    - 양방향 암호는 비인가된 사용자로부터 데이터 통신을 보호받기 위해 사용하는 암호 기법
        - 암호화한 데이터를 다시 복호화하면 데이터가 다시 원래 데이터로 돌아오게 되는 방법으로 양방향 암호와에서는 
        **암호화키, 복호화키** 라는 특정한 데이터를 요구하게 된다.

4. 데이터를 암호화하면, 인가된 사람만이 해당 데이터를 해독할 수 있으며, 이를 통해 데이터의 기밀성, 무결성 및 인증을 보장한다.

## **2. 대칭키와 공개키**
   
   

- 양방향 암호는 암호문을 복호화 할 수 있도록 구현된 암호 알고리즘을 말한다.

- 양방향 암호 알고리즘은 주로 3DES와 AES가 잘 알려져 있다. 또한 RSA와 DSA등이 있다.
    - 이런 알고리즘들은 크게 대칭키 암호 알고리즘과 비대칭키(공개키) 암호 알고리즘으로 구분할 수 있다.

- 대칭키 암호화 방식(Symmetric-key Security Method): 암호화, 복호화에 사용하는 키가 동일하다.

- 비대칭키/공개키 암호화 방식(Asymmetric-key Security Method): 암호화 복호화에 사용하는 키가 서로 다르다.

#### 1. 대칭키 알고리즘
   
   
   
- 발신자와 수신자가 동일한 비밀키 사용
- 메시지 전달 경로가 길어지면 위험
- 비밀키가 유출되면 큰 문제가 발생함
- 하지만, 속도가 빠르다.

#### 2. 비대칭키(공개키) 알고리즘
   
   
   
- 공개키 : 누구나 사용할 수 있는 키
- 비공개키 : 개인이 가지고 있는 비밀키
- 공개키를 사용하여 암호화를 하고, 비공개키를 이용하면 비밀키를 가진 사람만이 볼 수 있음

- 비대칭키를 반대로 사용한다면 아래와 같은 상황으로 바뀜   
    
    -  비공개키를 가진 사람이 암호화, 그리고 암호화된 자료를 받은 사람이 공개키로 복호화했을 경우 올바른 대답이라면?
        - 인증서의 원리로 사용됨

## **3. 암호화 알고리즘**

#### 1. RSA 알고리즘
- RSA는 Ron Rivest, Adi Shamir, Leonard Adleman의 이름의 앞 글자를 따서 지음
- RSA는 공개키 암호 기반으로 두 개의 큰 소수를 사용하여 공개키와 비밀키를 생성하는 알고리즘이다.
- Key Generation, Key Distribution, Encryption, Decryption 4개의 단계를 가지는 알고리즘
   
   
   
1. Key Generation
- 두 개의 큰 소수 p와 q를 선택
- n = p * q 를 계산하여 공개키와 비밀키의 모듈러로 사용
- ϕ(n) = (p − 1) * (q − 1) 를 계산한다.
- 공개키 e를 선택한다. e는 1 < e < ϕ(n) 범위 내에서 ϕ(n)과 서로 소인 수
- 비밀키 d를 계산한다. d는 e * d ≡ 1 (mod ϕ(n))을 만족하는 값이다. 즉, e의 모듈러 역원이다.

2. Key Distribution
- 공개 키 (e, n) : 공개적으로 배포되어 암호화에 사용된다.
- 비밀 키 (d, n) : 비밀리에 유지되어 복호화에 사용된다.

3. Encryption
- 평문 메세지 M을 암호화하기 위해, C = M^e(mod n)을 계산하여 암호문 C를 생성

4. Decrption
- 암호문 C를 복호화하기 위해, M = C^d(mod n)을 계산하여 평문 메세지 M을 복원

5. 예시
    1. 두 소수 p = 61, q = 53
    2. n = 61 * 53 = 3233
    3. ϕ(n) = (61-1) * (53-1) = 3120
    4. e = 17을 선택 (17은 1과 3120 사이의 소수이며, 3120과 서로 소)
    5. d = 2753을 계산 (17과 2753은 모듈러 3120에 대해 역원 관계에 있다.)
    6. 공개키 : (17, 3233), 비밀키 : (2753, 3233)
    7. 암호화 -> 평문 메세지 M = 65, C = 65^17*(mod 3233) = 2790
    8. 복호화 -> 암호문 C = 2790, M = 2790^2753*(mod 3233) = 65
